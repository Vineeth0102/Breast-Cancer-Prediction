```python
import numpy as np
import pandas as pd
```

```python
test = pd.read_csv(r"/content/can_data.csv")
test.info()
```

```python
test['diagnosis'].unique()
```

```python
df = test.dropna(axis = 1)
```

```python
df.info()
```

```python
df = df.drop('id', axis = 1)
```

```python
df['diagnosis'].value_counts()
```

```python
df.describe()
```

```python
df['diagnosis'] = df['diagnosis'].map({"M" : 1, "B" : 0})
```

```python
df.info()
df
```

```python
x = df.drop('diagnosis', axis = 1)
y = df['diagnosis']
```

```python
type(y)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.2,random_state=42)
```

```python
x_train.shape
```

```python
from sklearn.preprocessing import StandardScaler
ss = StandardScaler()
x_train = ss.fit_transform(x_train)
x_test = ss.fit_transform(x_test)
```

```python
from sklearn.linear_model import LogisticRegression

lr = LogisticRegression()
lr.fit(x_train,y_train)
```

```python
lr_pred = lr.predict(x_test)
lr_pred
```

```python
print(y_test)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
print(accuracy_score(y_test,lr_pred))
cm = confusion_matrix(y_test,lr_pred)
print(cm)
```

```python
sns.heatmap(cm, annot = True , fmt ='d')
plt.xlabel("Prediction")
plt.ylabel("True Value")
```

```python
lr_acc = accuracy_score(y_test,lr_pred)
```

```python
res = pd.DataFrame()
res
```

```python
temp_res = pd.DataFrame({'Algorithum' :['Logistic Regression Method'],'Accuracy' : [lr_acc]})
res = pd.concat([res, temp_res])
res[['Algorithum','Accuracy']]
```

```python
from sklearn.tree import DecisionTreeClassifier
Decision_class = DecisionTreeClassifier()
Decision_class.fit(x_train,y_train)
```

```python
y_pred_dr = Decision_class.predict(x_test)
y_pred_dr
dr_acc = accuracy_score(y_test,lr_pred)
```

```python
sns.heatmap(cm, annot = True , fmt ='d')
plt.xlabel("Prediction")
plt.ylabel("True Value")
```

```python
temp_res = pd.DataFrame({'Algorithum' :['Decision Tree Classifier'],'Accuracy' : [dr_acc]})
res = pd.concat([res, temp_res])
res[['Algorithum','Accuracy']]
```
